```python
termoyopal = pd.read_excel(r'TERMOYOPAL.xlsx')
dfy = pd.DataFrame(termoyopal)
dfy.head()
```

```python
plt.figure(figsize=(12,8))
plt.plot(dfy['Precio de Oferta Ideal $/kWh'], color = 'red',label = 'Precio')
plt.xticks(fontsize = 10)
plt.show()
```

```python
def TestStationaryPlot(ts):
    rolling_mean = ts.rolling(window = 12, center = False).mean()
    rolling_std = ts.rolling(window = 12, center = False).std()
    
    plt.figure(figsize=(12,8))
    plt.plot(ts, color = 'blue',label = 'Datos Originales')
    plt.plot(rolling_mean, color = 'red', label = 'Media Móvil')
    plt.plot(rolling_std, color ='black', label = 'Desviación Estándar')
    plt.xticks(fontsize = 10)
    plt.yticks(fontsize = 10)
    
    plt.xlabel('Dia', fontsize = 10)
    plt.ylabel('Precio', fontsize = 10)
    plt.legend(loc='Mejor', fontsize = 10)
    plt.title('Media Móvil & Desviación Estándar', fontsize = 15)
    plt.show(block= True)
    
def TestStationaryAdfuller(ts, cutoff = 0.01):
    ts_test = adfuller(ts, autolag = 'AIC')
    ts_test_output = pd.Series(ts_test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    
    for key,value in ts_test[4].items():
        ts_test_output['Critical Value (%s)'%key] = value
    print(ts_test_output)
    
    if ts_test[1] <= cutoff:
        print("Fuerte evidencia contra la hipótesis nula, rechaza la hipótesis nula. Los datos no tienen raíz unitaria, por lo tanto son estacionarios")
    else:
        print("Evidencia débil contra la hipótesis nula, las series temporales tienen una raíz unitaria, lo que indica que no es estacionaria") 
```

```python
Dia = dfy['Dia']
PrecioMedio =dfy['Precio de Oferta Ideal $/kWh']
TestStationaryPlot(PrecioMedio)
```

```python
from statsmodels.tsa.stattools import acf, pacf, adfuller, coint
TestStationaryAdfuller(PrecioMedio)
```

```python
Preciosln = np.log(dfy['Precio Deflactado'])
plt.figure(figsize=(12,8))
plt.plot(Preciosln, color = 'blue',label = 'Precio Deflactado')
plt.xticks(fontsize = 10)
plt.show()
```

```python
TestStationaryPlot(Preciosln)
```

```python
TestStationaryAdfuller(Preciosln)
```

```python
Preciosdiff = np.diff(Preciosln)
plt.figure(figsize=(12,8))
plt.plot(Preciosdiff, color = 'blue',label = 'Precio Diferenciado')
plt.xticks(fontsize = 10)
plt.show()
```

```python
Preciosdiffd = pd.DataFrame(Preciosdiff)
Preciosdiffd.dropna(inplace=True)
TestStationaryPlot(Preciosdiffd)
```

```python
#TestStationaryAdfuller(Preciosdiffd)
```

```python
Dia = dfy['Dia']
Precio = pd.DataFrame(Preciosln)
datos = pd.concat([Dia,Precio], axis=1)
```

```python
Entrenamiento = datos.loc[0:454]
Prueba = datos.loc[455:]
Entrenamiento.columns=['Dia','Precio']
Prueba.columns=['Dia','Precio']
Entrenamiento.index = pd.to_datetime(Entrenamiento['Dia'])
Prueba.index = pd.to_datetime(Prueba['Dia'])
del Entrenamiento['Dia']
del Prueba['Dia']
Entrenamiento = Entrenamiento[5:]
```

```python
plt.figure(figsize=(12,8))
plt.plot(Entrenamiento, color = 'red',label = 'Train')
plt.plot(Prueba, color = 'blue', label = 'Test')
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
    
plt.xlabel('Dia', fontsize = 10)
plt.ylabel('Precio', fontsize = 10)
plt.legend(loc='Mejor', fontsize = 10)
plt.title('Entrenamiento & Prueba', fontsize = 15)
plt.show(block= True)
```

```python
import statsmodels.api as sm
fig = plt.figure(figsize = (12,8))
ax1 = fig.add_subplot (211)
fig = sm.graphics.tsa.plot_acf(Preciosdiff, lags = 23, ax = ax1)
ax2 = fig.add_subplot (212)
fig = sm.graphics.tsa.plot_pacf(Preciosdiff, lags = 23, ax = ax2)
plt.show ()
```

```python
data = pd.concat([Entrenamiento, Prueba])
data.head(5)
```

```python
data.plot(figsize=(12,10))
plt.show()
```

```python
import itertools
p = range(0, 4) 
d = range(0, 1) 
q = range(0, 4) 
pdq = list(itertools.product(p, d, q)) 
pdq_x_QDQs = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p, d, q))] 
```

```python
for param in pdq:
    for seasonal_param in pdq_x_QDQs:
        try:
            mod = sm.tsa.statespace.SARIMAX(data,
                                            order=param,
                                            seasonal_order=seasonal_param,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{} - AIC:{}'.format(param, seasonal_param, results.aic))
        except:
            continue
```

```python
y_Prueba = Prueba.copy()
sarima = sm.tsa.statespace.SARIMAX(data, order=(2, 1, 3),seasonal_order=(1, 0, 2, 7)).fit()
y_Prueba['SARIMA'] = sarima.predict(start="2019-04-01", end="2019-06-30", dynamic=True)
plt.figure(figsize=(12,8))
plt.plot( Entrenamiento['Precio'], label='Entrenamiento')
plt.plot(Prueba['Precio'], label='Prueba')
plt.plot(y_Prueba['SARIMA'], label='SARIMA')
plt.legend(loc='best')
plt.show()
```

```python
Prediccion = sarima.forecast(7)
Prediccion.columns = ('Dia', 'Precio')
np.exp(Prediccion)
```

```python
sarima.summary()
```

```python
plt.figure(figsize=(12,8))
sarima.resid.plot(kind='kde')
plt.show()
```

```python
sarima.plot_diagnostics(figsize=(15, 12))
plt.show()
```
